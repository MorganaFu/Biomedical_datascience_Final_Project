**Add a "BMI (kg/m^2)" column to the dataframe by calculating body mass index based on weight and height provided** `BMI = weight(kg)/(height(m)**2)` **where** '`**`' **is  the exponent operator in Python. Be wary of units. (6 points)**

````python
weight = df_new['Weight (kg)'].tolist()
height = df_new['Height (cm)'].tolist()

# Initialize an empty list to store BMI values
BMI_values = []

for item in range(len(df_new)):
    BMI = weight[item]/((height[item]/100)**2)
    BMI_values.append(BMI)

# Assign the list of calculated BMI values to the DataFrame
df_new['BMI (kg/m^2)'] = BMI_values

df_new.head()
````

What statistical test should be employed to analyze the effectiveness of the new heart failure medication formulations developed by two pharmaceutical companies, sacubitril and enalapril, with respect to their claims of reducing NT-proBNP levels? The claims state that sacubitril reduces NT-proBNP levels by an average of 3000 within 4 weeks and 2000 within 8 weeks, while enalapril reduces NT-proBNP levels by an average of 2500 within 4 weeks and 1500 within 8 weeks. A clinical trial was conducted using samples from both companies, and the objective is to determine which formulation is more effective.

Clearly state your hypotheses for the combinations of analyses and do the significance test accordingly:

<font color='orange'>***1. For the sacubitril group, is there a significant difference between pre-treatment and Week 4 NT-proBNP levels?*** (2 Points)

From the actual meaning of the clinical trial, we can know that pre-treatment and Week 4 NT-proBNP levels are not independent samples; they are paired-samples. Thus, we should apply either Paired t-test or Wilcoxon Signed-rank test.

H0: There is no significant difference between the mean of pre-treatment(Î¼1) and Week 4 NT-proBNP(Î¼2) levels after treated with sacubitril.

$$ H_0: Î¼_1 = Î¼_2 $$

H1: There is significant difference between the mean of pre-treatment(Î¼1) and Week 4 NT-proBNP(Î¼2) levels after treated with sacubitril.

$$ H_1: Î¼_1 â‰  Î¼_2 $$

<font color='yellow'>**Let us figure out whether two measurements(pre-treatment and Week 4 NT-proBNP levels) follow normal distribution at first.**

````python
# Since the sample size is larger than 50, choose Kolmogorov-Smirnov Test to see if they follow the normal distribution
# Also, we should notice that since they are related samples rather than independent samples, we should apply one-sample KS test
import scipy.stats as stats
pre_treatment_sacubitril= pre_treatment.loc[pre_treatment['Group']=='Sacubitril-valsartan','Pre-treatment NT-proBNP (pg/mL)']
week_4_NT_proBNP_sacubitril= week4.loc[week4['Group']=='Sacubitril-valsartan','Week 4 NT-proBNP (pg/mL)']
````

````python
pre_treatment_sacubitril
````

````python
week_4_NT_proBNP_sacubitril
````

````python
# Perform one-sample KS test for pre-treatment data to see if it follows normal distribution
ks_statistic_pre, p_value_pre = stats.kstest(pre_treatment_sacubitril, 'norm', args=(pre_treatment_sacubitril.mean(), pre_treatment_sacubitril.std()))
print(f"Pre-treatment KS Test: KS Statistic = {ks_statistic_pre}, P-value = {p_value_pre}")

# Perform one-sample KS test for Week 4 data to see if it follows normal distribution
ks_statistic_week4, p_value_week4 = stats.kstest(week_4_NT_proBNP_sacubitril, 'norm', args=(week_4_NT_proBNP_sacubitril.mean(), week_4_NT_proBNP_sacubitril.std()))
print(f"Week 4 KS Test: KS Statistic = {ks_statistic_week4}, P-value = {p_value_week4}")
````

From the result of KS test shows that both of the P-value close to 0, which means that observed data is unlikely fall into the H0 region(follow normal distribution). The conlusion is that they do not follow the normal distribution.

**Thus, we need to apply Wilcoxon Signed-rank test.**


````python
w_statistics, p_value = stats.wilcoxon(pre_treatment_sacubitril, week_4_NT_proBNP_sacubitril)
print(f"Wilcoxon signed-rank test statistic: {w_statistics}, P-value: {p_value}")
````

From the result of Wilcoxon signed-rank we know that our P-value is close to 0, which means that it is unlikely to observe result under the hypothesis - the mean of pre-treatment and week 4 NT-proBNP is the same. <font color='lightgreen'>Thus, the conclusion is that there is a significant difference of pre-treatment NT-proBNP level and after Sacubitril-valsartan was given for 4 weeks.

From the actual meaning of the clinical trial, we could know that Enalapril and Sacubitril-valsartan groups are mutually exclusive. Thus, we should either apply independent t-test or Welch's t-test.

<font color='yellow'> **Let us figure out whether two samples(Week 4 NT-proBNP level treated with enalapril and Week 4 NT-proBNP level treated with Sacubitril-valsartan) follow the same distribution at first.**

````python
import scipy.stats as stats
week4_enalapril = week4.loc[week4['Group']=='Enalapril','Week 4 NT-proBNP (pg/mL)']
week4_sacubitril = week4.loc[week4['Group']=='Sacubitril-valsartan','Week 4 NT-proBNP (pg/mL)']
# Two-sample KS test (comparing data1 and data2)
ks_statistic_week4, p_value_week4 = stats.ks_2samp(week4_enalapril, week4_sacubitril)
print("Two-sample KS test statistic:", ks_statistic_week4, "P-value:", p_value_week4)
````

From the result of two-sample KS test, we could know that the P-value is quite small, indicating that it is unlikely to observe any data under the H0- two samples follow the same distribution.<font color='pink'> **Thus, to make our research rigorous, we need to apply Mann-Whitney U test to see whether there is a difference in the central tendency(median) between two samples.**


H0: There is no significant difference between the median of Week 4 NT-proBNP group where patients were given Enalapril(xÍ‚1) and Week 4 NT-proBNP group where patients were given Sacubitril-valsartan(xÍ‚2).

$$ H_1: xÍ‚_1 â‰  xÍ‚_2 $$

````python
# Apply Mann-Whitney U test
u_statistic, p_value = stats.mannwhitneyu(week4_enalapril, week4_sacubitril, alternative='two-sided')
print("Mann-Whitney U statistic:", u_statistic, "P-value:", p_value)
````

From the result of the Mann-Whitney U test, we can know that it rejects H0.<font color='lightgreen'> Thus, there is a significance different between two medians(Week 4 NT-proBNP level treated with enalapril and Week 4 NT-proBNP level treated with Sacubitril-valsartan).
